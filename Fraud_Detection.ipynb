## Outliers Treatment:

As the data set is already PCA transformed, we will use deletion of outliers as the Outliers Treatment process. Attribute V4 does not have any outliers in terms for fraud cases, but V11, V10, V12, and V14 have outliers. 

```python
V11_fraud=bal_card_data[bal_card_data['Class']==1]['V11']
```

```python
V11_q25,V11_q75=np.percentile(V11_fraud,25), np.percentile(V11_fraud,75)
V11_IQR=V11_q75-V11_q25
V11_lower,V11_upper=V11_q25-1.5*V11_IQR,V11_q75+1.5*V11_IQR
```

```python
bal_card_data = bal_card_data.drop(bal_card_data[(bal_card_data['V11'] > V11_upper) | (bal_card_data['V11'] < V11_lower)].index)
```

```python
sns.boxplot(data=bal_card_data,x="Class",y='V11')
```

```python
#outlier detection for V10, V12 and V14:

# fraud data for each attribute: 
V10_fraud=bal_card_data[bal_card_data['Class']==1]['V10']
V12_fraud=bal_card_data[bal_card_data['Class']==1]['V12']
V14_fraud=bal_card_data[bal_card_data['Class']==1]['V14']

#Q25 and Q75 for each attribute:
V10_q25,V10_q75=np.percentile(V10_fraud,25), np.percentile(V10_fraud,75)
V12_q25,V12_q75=np.percentile(V12_fraud,25), np.percentile(V12_fraud,75)
V14_q25,V14_q75=np.percentile(V14_fraud,25), np.percentile(V14_fraud,75)

#IQR for each attribute:
V10_IQR=V10_q75-V10_q25
V12_IQR=V12_q75-V12_q25
V14_IQR=V14_q75-V14_q25

#upper and lower extreams for each attribute:
V10_lower,V10_upper=V10_q25-1.5*V10_IQR,V10_q75+1.5*V10_IQR
V12_lower,V12_upper=V12_q25-1.5*V12_IQR,V12_q75+1.5*V12_IQR
V14_lower,V14_upper=V14_q25-1.5*V14_IQR,V14_q75+1.5*V14_IQR

#Removing outliers for each attribute:
bal_card_data = bal_card_data.drop(bal_card_data[(bal_card_data['V10'] > V10_upper) | 
                                   (bal_card_data['V10'] < V10_lower)].index)
bal_card_data = bal_card_data.drop(bal_card_data[(bal_card_data['V12'] > V12_upper) | 
                                   (bal_card_data['V12'] < V12_lower)].index)
bal_card_data = bal_card_data.drop(bal_card_data[(bal_card_data['V14'] > V14_upper) | 
                                   (bal_card_data['V14'] < V14_lower)].index)
```

For spliting the data we will use Stratified k-fold cross validation, so that we can test the model on the real imbalanced data. Stratified k-fold will guarantees the proportion of fraud transctions in the test data set will same as the proportion of fraud transctions in complete data set. 

```python
from sklearn.model_selection import StratifiedKFold
X_org=bal_card_data.drop('Class',axis=1)
Y_org=bal_card_data['Class']

skf = StratifiedKFold(n_splits=5,random_state=42,shuffle=True)

for train_index, test_index in skf.split(X_org,Y_org):
  X_org_train, X_org_test = X_org.iloc[train_index], X_org.iloc[test_index]
  Y_org_train, Y_org_test = Y_org.iloc[train_index],Y_org.iloc[test_index]
```
